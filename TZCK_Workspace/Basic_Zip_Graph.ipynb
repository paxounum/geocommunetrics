```python
#Establishing the Graph output for linear, upright inputs as "zip_up" variable
```

```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
plt.plot(placement,zip_up)
plt.title("Kansas City: 64128")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
#Output for "upside-down" inputs, pre-calculated as "zip_down" variable
```

```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
zip_down = [3, 4, 5, 2, 5]
plt.plot(placement,zip_up)
plt.plot(placement,zip_down)
plt.title("Kansas City: 64128")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
#zip_up and zip_down can be graphed at once
#then both lists must be reversed, and their mapping axese flipped to map zip_right and zip_left respectively
#I will now try to do them in succession by resetting UP and DOWN
#In order to set RIGHT and LEFT as revered items, I used [::-1] to call a list in reverse to define the variable
```

```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
zip_down = [3, 4, 5, 2, 5]
```

```python
zip_up[:]
```

```python
zip_right = zip_up[::-1]
```

```python
print(zip_right)
```

```python
zip_left = zip_down[::-1]
print(zip_left)
```

```python
print(zip_up, zip_right, zip_down, zip_left)
```

```python
plt.plot(placement,zip_up)
plt.plot(placement,zip_down)
plt.plot(zip_right, placement)
plt.plot(zip_left, placement)
```
