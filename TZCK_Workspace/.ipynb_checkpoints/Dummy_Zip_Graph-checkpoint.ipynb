```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
plt.plot(placement,zip_up)
plt.title("Kansas City: 64128")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
#Output for "upside-down" inputs, pre-calculated as "zip_down" variable
```

```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
zip_down = [3, 4, 5, 2, 5]
plt.plot(placement,zip_up)
plt.plot(placement,zip_down)
plt.title("Kansas City: 64128")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
#zip_up and zip_down can be graphed at once
#then both lists must be reversed, and their mapping axese flipped to map zip_right and zip_left respectively
#I will now try to do them in succession by resetting UP and DOWN
#In order to set RIGHT and LEFT as revered items, I used [::-1] to call a list in reverse to define the variable
```

```python
placement = [1, 2, 3, 4, 5]
zip_up = [1, 4, 1, 2, 3]
zip_down = [3, 4, 5, 2, 5]
```

```python
zip_up[:]
```

```python
zip_right = zip_up[::-1]
```

```python
print(zip_right)
```

```python
zip_left = zip_down[::-1]
print(zip_left)
```

```python
print(zip_up, zip_right, zip_down, zip_left)
```

```python
#blue
plt.plot(placement,zip_up)
#orange
plt.plot(placement,zip_down)
#green
plt.plot(zip_right, placement)
#red
plt.plot(zip_left, placement)
```

```python
#Now let's play with color and style!
```

```python
plt.plot(placement,zip_up, color = 'k', lw = 5)
plt.plot(placement,zip_down, color = 'k', lw = 5)
plt.plot(zip_right, placement,color = 'k', lw = 5)
plt.plot(zip_left, placement,color = 'k', lw = 5)
plt.axis('off')
```

```python
#drawing the 2x2 subplots
fig, axs = plt.subplots(ncols=2, nrows=2)

#assigning data to those subplots
#turning the axis off

axs[0,0].plot(placement,zip_up, color = 'g', lw = 5)
axs[0,0].plot(placement,zip_down, color = 'g', lw = 5)
axs[0,0].plot(zip_right, placement,color = 'g', lw = 5)
axs[0,0].plot(zip_left, placement,color = 'g', lw = 5)
axs[0,0].axis('off')

#reversed the zip order so that shape would be mirrored
axs[0,1].plot(placement,zip_up[::-1], color = 'b', lw = 5)
axs[0,1].plot(placement,zip_down[::-1], color = 'b', lw = 5)
axs[0,1].plot(zip_right[::-1], placement,color = 'b', lw = 5)
axs[0,1].plot(zip_left[::-1], placement,color = 'b', lw = 5)
axs[0,1].axis('off')

axs[1,0].plot(placement,zip_up[::-1], color = 'r', lw = 5)
axs[1,0].plot(placement,zip_down[::-1], color = 'r', lw = 5)
axs[1,0].plot(zip_right[::-1], placement,color = 'r', lw = 5)
axs[1,0].plot(zip_left[::-1], placement,color = 'r', lw = 5)
axs[1,0].axis('off')

axs[1,1].plot(placement,zip_up, color = 'y', lw = 5)
axs[1,1].plot(placement,zip_down, color = 'y', lw = 5)
axs[1,1].plot(zip_right, placement,color = 'y', lw = 5)
axs[1,1].plot(zip_left, placement,color = 'y', lw = 5)
axs[1,1].axis('off')

fig.tight_layout()

#eliminate space between plots
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=-.09, hspace=-.09)
plt.show()
```

#### how to systematize a larger output (e.g 4x4)?

```python


#determing the HxW of square grid
grid = 2

#drawing the 2x2 subplots
fig, axs = plt.subplots(ncols=grid, nrows=grid)


#determining grid place variables
row = 0
column = 0

#assigning data to those subplots
#turning the axis off

#Original 
#def original():
axs[row,column].plot(placement,zip_up, color = 'g', lw = 5)
axs[row,column].plot(placement,zip_down, color = 'g', lw = 5)
axs[row,column].plot(zip_right, placement,color = 'g', lw = 5)
axs[row,column].plot(zip_left, placement,color = 'g', lw = 5)
axs[row,column].axis('off')

#Mirror
#def mirror():
axs[row,column+1].plot(placement,zip_right, color = 'b', lw = 5)
axs[row,column+1].plot(placement,zip_left, color = 'b', lw = 5)
axs[row,column+1].plot(zip_up, placement,color = 'b', lw = 5)
axs[row,column+1].plot(zip_down, placement,color = 'b', lw = 5)
axs[row,column+1].axis('off')

#Mirror
#def mirror():
axs[row+1,column].plot(placement,zip_right, color = 'b', lw = 5)
axs[row+1,column].plot(placement,zip_left, color = 'b', lw = 5)
axs[row+1,column].plot(zip_up, placement,color = 'b', lw = 5)
axs[row+1,column].plot(zip_down, placement,color = 'b', lw = 5)
axs[row+1,column].axis('off')

#Original 
#def original():
axs[row+1,column+1].plot(placement,zip_up, color = 'g', lw = 5)
axs[row+1,column+1].plot(placement,zip_down, color = 'g', lw = 5)
axs[row+1,column+1].plot(zip_right, placement,color = 'g', lw = 5)
axs[row+1,column+1].plot(zip_left, placement,color = 'g', lw = 5)
axs[row+1,column+1].axis('off')



#original()
#mirror()
fig.tight_layout()

#eliminate space between plots
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=-.09, hspace=-.09)
plt.show()
```

```python
# how to add rows and columns after plotting
import numpy as np

def textile():
    grid = 4
    fig, axs = plt.subplots(ncols=grid, nrows=grid)
    col = 0
    row = 0
    line = 0
    for rep in range (1,grid):
        for col in range (0,grid):
            axs[row,col].plot(placement,zip_up, color = 'g', lw = 3)
            axs[row,col].plot(placement,zip_down, color = 'g', lw = 3)
            axs[row,col].plot(zip_right, placement,color = 'g', lw = 3)
            axs[row,col].plot(zip_left, placement,color = 'g', lw = 3)
            axs[row,col].axis('off')
        row += 1
    for col in range (0,grid):
        axs[row,col].plot(placement,zip_up, color = 'g', lw = 3)
        axs[row,col].plot(placement,zip_down, color = 'g', lw = 3)
        axs[row,col].plot(zip_right, placement,color = 'g', lw = 3)
        axs[row,col].plot(zip_left, placement,color = 'g', lw = 3)
        axs[row,col].axis('off')
    row += 1
        
textile()
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=-.09, hspace=-.09)
plt.show()
```
