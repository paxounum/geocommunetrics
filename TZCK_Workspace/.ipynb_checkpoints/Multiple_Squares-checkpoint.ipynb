```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(ncols=3, nrows=3)
gs = axs[1, 2].get_gridspec()

fig.tight_layout()

#eliminating space between the subplots
# from: https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=0)
plt.show()
```

```python
#elimintating the axes for each box
#to show this, a standard chart of sin will be used as the plot

import numpy as np

#establish variables to be graphed
x=np.linspace(-3,3,100)
y1=np.sin(x)

plt.plot(x, y1)

plt.show()
```

```python
#drawing the 4x4 subplots
fig, axs = plt.subplots(ncols=2, nrows=2)

#assigning data to those subplots

axs[0,0].plot(x, y1)
axs[0,1].plot(x, y1)
axs[1,0].plot(x, y1)
axs[1,1].plot(x, y1)

fig.tight_layout()

#eliminate space between plots
#from: https://www.delftstack.com/howto/matplotlib/how-to-turn-off-the-axes-for-subplots-in-matplotlib/

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=0)
plt.show()
```

```python
#drawing the 4x4 subplots
fig, axs = plt.subplots(ncols=2, nrows=2)

#assigning data to those subplots
#turning the axis off

axs[0,0].plot(x, y1)
axs[0,0].axis('off')

axs[0,1].plot(x, y1)
axs[0,1].axis('off')

axs[1,0].plot(x, y1)
axs[1,0].axis('off')

axs[1,1].plot(x, y1)
axs[1,1].axis('off')

fig.tight_layout()

#eliminate space between plots
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0, hspace=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()
```
